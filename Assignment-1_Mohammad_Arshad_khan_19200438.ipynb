# Task 1: Algorithm to compute the factorial of an integer "n"

### Pseudocode for Factorial of a number :

* Declare N and F as integer variable.  
* Initialize F=1.
* Enter the value of N.  
* Check whether N>0, if not then F=1.  
* If yes then, F=F*N  
* Decrease the value of N by 1 .  
* Repeat step 4 and 5 until N=0.
* Now print the value of F.  

The final value of F will be the factorial of N(number).

## None-Tail Recursion- Factorial

```python
# Recursive non-tail method for computing factorial
# Here in recursion the function call itself and every call has a unique context
# (own values for variables,parameters and own memory address), therefore iterative is better 
# in terms of storage and time effeciency
# below is a non-tail recursion computing the factorial
# It's none-tail because we have return type after every call

def factorial_non_tail(n):
    if n == 1: # this is the stopping condition (base case), stopping condition is a base case
        return 1
    else:
        return n * factorial_non_tail(n-1) # We are multiplying the return value by n

```

## Tail Recursion -Factorial

```python
# tail recursion
# The tail recursion is better than non-tail recursion. 
# As there is no task left after the recursive call, it will be easier for the compiler to optimize the code. 
# When one function is called, its address is stored inside the stack.

def factorial_tail(n, accumulator):
    if n == 1:
        return accumulator
    else:
        return factorial_tail(n-1, n*accumulator)
```

## Task 3: Linear search , iterative binary search and recursive binary search.

### Linear Search Algorithm (Arr, x)

* Set i to 1
* if i > len(arr) then go to step 7
* if Arr[i] = x then go to step 6
* Set i to i + 1
* Go to Step 2
* Print Element x Found at index i and go to step 8
* Print element not found
* Exit

### Binary Search Algorithm 

* A ← sorted array
* n ← size of array
* x ← value to be searched

* Set lowerBound = 1
* Set upperBound = n 

* while x not found
      if upperBound < lowerBound 
         EXIT: x does not exists.
   
* set midPoint = lowerBound + ( upperBound - lowerBound ) / 2
      
* if A[midPoint] < x
         set lowerBound = midPoint + 1
         
* if A[midPoint] > x
         set upperBound = midPoint - 1 

* if A[midPoint] = x 
         EXIT: x found at location midPoint
* end while
   
* end procedure


Ref :- https://www.tutorialspoint.com/data_structures_algorithms/binary_search_algorithm.htm

Note: Instead of using the factorail of elements as elements in list, I have used the "n" as elements.

```python
# Linear Search

def search(arr, x): 
  
    for i in range(len(arr)): 
  
        if arr[i] == x: 
            return i 
  
    return -1
```

```python
# Binary Search using recursion

def binary_search_recursive(arr, elem, start=0, end=None):
    if end is None:
        end = len(arr) - 1
    if start > end:
        return False

    mid = (start + end) // 2
    if elem == arr[mid]:
        return mid
    if elem < arr[mid]:
        return binary_search_recursive(arr, elem, start, mid-1)
    # elem > arr[mid]
    return binary_search_recursive(arr, elem, mid+1, end)
```

```python
# Binary search using iteration

def binary_search_iterative(arr, elem):
    start, end = 0, (len(arr) - 1)
    while start <= end:
        mid = (start + end) // 2
        if elem == arr[mid]:
            return mid
        if elem < arr[mid]:
            end = mid - 1
        else:  # elem > arr[mid]
            start = mid + 1

    return False
```
